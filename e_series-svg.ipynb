```python
import math
import cmath
import drawSvg as draw
```

```python
scale = cmath.pi
```

```python
N = 50
args = list(map(lambda x: complex(0, x/N*scale), range(N)))
points = list(map(lambda z: cmath.exp(z), args))
```

```python
def draw_points(pts, stroke='black'):
    p = draw.Path(stroke_width=0.01, stroke=stroke, fill='none')
    start = pts[0]
    p.M(start.real, start.imag)
    for z in pts[1:]:
        p.L(z.real, z.imag)
    return p
```

```python
d = draw.Drawing(8, 8, origin='center', displayInline=False)

d.append(draw_points(args, stroke='blue'))
d.append(draw_points(points, stroke='red'))

d.setRenderSize(w=800)
display(d)
```

```python
def cpow(z,p):
    if z == 0:
        return 0
    return cmath.exp(p*cmath.log(z))

def e_series(z, i):
    if i == 0:
        return 1
    else:
        return cpow(z, i) / math.factorial(i)
    
def sum_series(ps):
    sps = []
    s = 0j
    for pp in ps:
        s += pp
        sps.append(s)

    return sps
```

```python
d = draw.Drawing(8, 8, origin='center', displayInline=False)

d.append(draw_points(points, stroke='blue'))
for p in args:
    ps = list(map(lambda i: e_series(p,i), range(20)))
    sps = sum_series(ps)
    
    d.append(draw_points(sps, stroke='red'))

d.setRenderSize(w=800)
display(d)
```
