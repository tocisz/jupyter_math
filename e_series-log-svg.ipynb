```python
import math
import cmath
import drawSvg as draw
```

```python
scale = 1 #cmath.pi
```

```python
N = 64
points = list(map(lambda x: complex(0.5, (x/N)*scale), range(N)))
args = list(map(lambda z: cmath.log(z), points))
```

```python
def draw_points(d, pts, stroke='black'):
    px_size = d.width/d.renderWidth
    p = draw.Path(stroke_width=px_size, stroke=stroke, fill='none')
    start = pts[0]
    p.M(start.real, start.imag)
    for z in pts[1:]:
        p.L(z.real, z.imag)
    d.append(p)
```

```python
d = draw.Drawing(2, 3, origin=(-1,-1), displayInline=False)
d.setRenderSize(w=800)

draw_points(d, args, stroke='blue')
draw_points(d, points, stroke='red')

display(d)
```

```python
def cpow(z,p):
    if z == 0:
        return 0
    return cmath.exp(p*cmath.log(z))

def e_series(z, i):
    if i == 0:
        return 1
    else:
        return cpow(z, i) / math.factorial(i)
    
def sum_series(ps):
    sps = []
    s = 0j
    for pp in ps:
        s += pp
        sps.append(s)

    return sps
```

```python
d = draw.Drawing(2, 3, origin=(-0.5,-1), displayInline=False)
d.setRenderSize(w=800)

draw_points(d, points, stroke='blue')
for p in args:
    ps = list(map(lambda i: e_series(p,i), range(20)))
    sps = sum_series(ps)
    
    draw_points(d, sps, stroke='red')

display(d)
```
